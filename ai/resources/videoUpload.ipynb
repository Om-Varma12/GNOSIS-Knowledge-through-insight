```python
import os
import uuid
import torch
import clip
import whisper
import cv2
import numpy as np
from tqdm import tqdm
from qdrant_client import QdrantClient
from qdrant_client.models import PointStruct
from dotenv import load_dotenv
import pandas as pd
from PIL import Image
import re
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

# CLIP
clip_model, preprocess = clip.load("ViT-B/32", device=device)
clip_model.eval()

# Whisper (multilingual, handles Marathi, Hindi, English, etc)
whisper_model = whisper.load_model("base")  # or "small" if you want better quality
```

```python
load_dotenv()

QDRANT_KEY = os.getenv("QDRANT_KEY")
QDRANT_URL = os.getenv("QDRANT_URL")

client = QdrantClient(
    url=QDRANT_URL,
    api_key=QDRANT_KEY
)

COLLECTION_NAME = "GNOSIS_V2"
```

```python
BASE_VIDEO_ROOT = r"D:/STUDY/PROJECTS/GNOSIS/Resources/Videos"

DOMAINS = {
    "medical": {
        "folder": "Medical",
        "excel": "medicalURL.xlsx"
    },
    "politics": {
        "folder": "Politics",
        "excel": "politicsURL.xlsx"
    }
}

BATCH_SIZE = 4
UPLOAD_BATCH = 16
FRAMES_PER_VIDEO = 8
```

```python
def extract_frames(video_path, num_frames=8):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return []

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames <= 0:
        cap.release()
        return []

    idxs = np.linspace(0, total_frames - 1, num_frames).astype(int)

    frames = []
    for i in idxs:
        cap.set(cv2.CAP_PROP_POS_FRAMES, int(i))
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(frame)

    cap.release()
    return frames
```

```python
def transcribe_video(video_path):
    try:
        result = whisper_model.transcribe(video_path)
        return result.get("text", "").strip()
    except Exception as e:
        print("âŒ Transcription failed:", video_path, e)
        return ""
```

```python
video_vec_batch = []
text_vec_batch = []
meta_batch = []
transcript_batch = []
points_buffer = []
```

```python
for domain_name, cfg in DOMAINS.items():

    folder = os.path.join(BASE_VIDEO_ROOT, cfg["folder"])
    excel_path = os.path.join(folder, cfg["excel"])

    print(f"\nðŸ“ Processing {domain_name.upper()}")

    if not os.path.exists(excel_path):
        print("âŒ Missing Excel:", excel_path)
        continue

    df = pd.read_excel(excel_path)

    print("Rows:", len(df))

    for idx, row in tqdm(df.iterrows(), total=len(df)):

        fname = str(row["file_name"]).strip()
        video_url = str(row["video_url"]).strip()
        label = str(row["label"]).strip()

        # Try common extensions
        possible_paths = [
            os.path.join(folder, fname),
            os.path.join(folder, fname + ".mp4"),
            os.path.join(folder, fname + ".mkv"),
            os.path.join(folder, fname + ".webm"),
        ]

        video_path = None
        for p in possible_paths:
            if os.path.exists(p):
                video_path = p
                break

        if video_path is None:
            print("âš ï¸ Missing video file for:", fname)
            continue

        # =========================
        # 1. Extract frames
        # =========================
        frames = extract_frames(video_path, FRAMES_PER_VIDEO)
        if not frames:
            continue

        frame_tensors = torch.stack([
            preprocess(Image.fromarray(f)) for f in frames
        ]).to(device)

        # =========================
        # 2. Encode video frames
        # =========================
        with torch.no_grad():
            frame_vecs = clip_model.encode_image(frame_tensors)
            frame_vecs = frame_vecs / frame_vecs.norm(dim=-1, keepdim=True)

        video_vec = frame_vecs.mean(dim=0).cpu().numpy()

        # =========================
        # 3. Transcribe
        # =========================
        transcript = transcribe_video(video_path)

        if transcript.strip():
            tokens = clip.tokenize([transcript], truncate=True).to(device)
            with torch.no_grad():
                text_vec = clip_model.encode_text(tokens).cpu().numpy()[0]
        else:
            text_vec = None

        # =========================
        # 4. Meta
        # =========================
        meta = {
            "modality": "video",
            "domain": domain_name,
            "dataset": "Video_Demo",
            "filename": os.path.basename(video_path),
            "video_path": video_path,
            "video_url": video_url,
            "label": label,
            "has_transcript": bool(transcript.strip()),
        }

        # =========================
        # 5. Batch
        # =========================
        video_vec_batch.append(video_vec)
        text_vec_batch.append(text_vec)
        meta_batch.append(meta)
        transcript_batch.append(transcript)

        # =========================
        # 6. Upload batch
        # =========================
        if len(video_vec_batch) >= BATCH_SIZE:

            for vvec, tvec, meta, trans in zip(video_vec_batch, text_vec_batch, meta_batch, transcript_batch):

                vectors_payload = {
                    "video": vvec.tolist()
                }

                if tvec is not None:
                    vectors_payload["text"] = tvec.tolist()

                point = PointStruct(
                    id=str(uuid.uuid4()),
                    vector=vectors_payload,
                    payload={
                        **meta,
                        "transcript": trans
                    }
                )

                points_buffer.append(point)

            video_vec_batch = []
            text_vec_batch = []
            meta_batch = []
            transcript_batch = []

            if len(points_buffer) >= UPLOAD_BATCH:
                client.upsert(collection_name=COLLECTION_NAME, points=points_buffer)
                print("â¬†ï¸ Uploaded", len(points_buffer), "video points")
                points_buffer = []
```

```python
if video_vec_batch:
    for vvec, tvec, meta, trans in zip(video_vec_batch, text_vec_batch, meta_batch, transcript_batch):

        vectors_payload = {
            "video": vvec.tolist()
        }

        if tvec is not None:
            vectors_payload["text"] = tvec.tolist()

        point = PointStruct(
            id=str(uuid.uuid4()),
            vector=vectors_payload,
            payload={
                **meta,
                "transcript": trans
            }
        )

        points_buffer.append(point)

if points_buffer:
    client.upsert(collection_name=COLLECTION_NAME, points=points_buffer)
    print("â¬†ï¸ Final upload:", len(points_buffer))

print("âœ… Video ingestion complete")
```

```python
import os
import numpy as np

# ===================== CONFIG =====================
QUERY = "indian boy Ashwamit Gautam got arrested"

TEXT_COLLECTION = "GNOSIS"
VIDEO_COLLECTION = "GNOSIS_V2"

# ===================== EMBED QUERY =====================
tokens = clip.tokenize([QUERY], truncate=True).to(device)

with torch.no_grad():
    qvec = clip_model.encode_text(tokens).cpu().numpy()[0]

qvec = qvec.tolist()   # â— DO NOT NORMALIZE (your working code didn't)

# ===================== SEARCH =====================

# ---- TEXT from GNOSIS ----
text_results = client.query_points(
    collection_name=TEXT_COLLECTION,
    query=qvec,
    using="text",
    limit=5,
    with_payload=True
)

# ---- IMAGES from GNOSIS (SEARCH USING VISION VECTOR) ----
image_results = client.query_points(
    collection_name=TEXT_COLLECTION,
    query=qvec,
    using="vision",   # âœ… THIS IS THE KEY FIX
    limit=5,
    with_payload=True
)

# ---- VIDEO from GNOSIS_V2 ----
video_results = client.query_points(
    collection_name=VIDEO_COLLECTION,
    query=qvec,
    using="video",
    limit=3,
    with_payload=True
)

# ===================== PRINT RESULTS =====================

# ---------- TEXT ----------
print("\n================ TEXT RESULTS ================\n")

for i, hit in enumerate(text_results.points, 1):
    payload = hit.payload

    print(f"#{i} SCORE={hit.score:.4f}")
    print("TITLE / FILE:", payload.get("title") or payload.get("filename"))
    print("URL:", payload.get("source_url") or payload.get("url"))
    print()


print("\n================ IMAGE RESULTS (Filtered: Screenshot only) ================\n")

img_count = 0

for hit in image_results.points:
    payload = hit.payload
    path = payload.get("source_path", "") or ""

    if "screenshot" not in path.lower():
        continue

    img_count += 1

    filename = os.path.basename(path)
    m = re.search(r"(\d+)\.png$", filename)
    image_number = m.group(1) if m else filename

    print(f"#{img_count} SCORE={hit.score:.4f}")
    print("IMAGE_ID:", image_number)   # âœ… ONLY: 193339
    print("TEXT:", payload.get("text"))
    print()

    if img_count >= 5:
        break





# ---------- VIDEO ----------
print("\n================ VIDEO RESULTS ================\n")

for i, hit in enumerate(video_results.points, 1):
    payload = hit.payload

    print(f"#{i} SCORE={hit.score:.4f}")
    print("FILE:", payload.get("filename"))
    print("URL:", payload.get("video_url"))
    print()
```
